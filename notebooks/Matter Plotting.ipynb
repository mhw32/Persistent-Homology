```python
import sys, numpy as np
import seaborn as sns
sns.set_style('whitegrid')
sys.path.append('../')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickorder = ['EC', 'Sil (d0)', 'Sil (1)', 'Sil (2)', 'Sil (P)', 'GC (0)', 'GC (1)', 'GC (2)', 'GKD (0)', 'GKD (1)', 'GKD (2)'] 
order = ['euler', 'silh_dim_1', 'silh_dim_2', 'silh_dim_3', 'all_silh', 'contour_dim_1', 'contour_dim_2', 'contour_dim_3', 'globalkde_dim_1', 'globalkde_dim_2', 'globalkde_dim_3']
split3 = np.exp([-26.14186, -17.15937, -7.790842, -0.2590126, -15.19706, -3.7297014, -2.4079456, -2.430418, -0.8822320, -0.7876049, -0.7023436])
split3norm = np.exp([-14.19302, -1.481107, -0.5755777, -1.771103, -1.151928, -0.3681693, 0, -0.04082199, -0.7984966, -0.6987211, -0.6634393])
split2 = np.exp([-3.140673, -8.967393, -2.519644, -0.04404427, -6.06745, -1.8838748, -0.6577800, -1.01060141, -0.6334591, -0.6339058, -0.5468125])
split2norm = np.exp([-2.476211, -1.108296, -0.4118814, -0.8685645, -0.2716508, -0.06827884, -0.16723592, -0.05340078, -0.5063470, -0.5339873, -0.5086138])
split4 = np.exp([-76.46603, -58.40512, -12.83022, -0.4000793, -np.inf, -np.inf, -6.2146081, -6.2146081, -1.0822552, -0.8152460, -0.7377763])
split4norm = np.exp([-55.83524, -10.83384, -1.303785, -2.792974, -12.8459, -1.599488, -0.2536028, -0.4034671, -1.0503746, -0.7109918, -0.7508386])
```

```python
import matplotlib
plt.figure(figsize=(20,5))
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18)
graph1, = plt.plot(np.arange(len(order)), split3, 'o', color='b', alpha=0.6, markersize=10, label='Triple Split')
graph2, = plt.plot(np.arange(len(order)), split2, 'v', color='r', alpha=0.6, markersize=10, label='Double Split')
graph3, = plt.plot(np.arange(len(order)), split4, 's', color='g', alpha=0.6, markersize=10, label='Quadruple Split')
plt.xticks(np.arange(len(order)), tickorder, fontsize=20)
plt.ylabel('p-value', fontsize=20)
plt.title('Comparison of Split Combinations on Hypothesis Tests on CDM/WDM Eagle Dataset.', fontsize=20)
plt.legend([graph1, graph2, graph3], fontsize=20)
plt.savefig("../saved_states/split-unnorm.png")
plt.savefig("../saved_states/split-unnorm.pdf")
plt.show()
```
